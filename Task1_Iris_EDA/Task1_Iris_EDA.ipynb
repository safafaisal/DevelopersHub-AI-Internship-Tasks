```python
Task 1: Exploring and Visualizing Iris Dataset

Goal:
To load the Iris dataset, explore its structure, and understand relationships between features using visualizations.
```

```python
#Import required libraries
import pandas as pd # for data handling
import seaborn as sns # for dataset and visualizaton
import matplotlib.pyplot as plt # for plotting graphs
```

```python
# Load dataset (source)
iris = sns.load_dataset("iris")

# Convert into pandas DataFrame 
df = pd.DataFrame(iris)
df.head() # Display first 5 rows of dataset
```

```python
df.shape #Check no of rows and columns
```

```python
df.columns # display column names
```

```python
df.info()
```

```python
df.describe() # Statistical summary of numerical columns
```

```python
# Scatter plot shows relationship between sepal length and sepal width
# Hue = species helps us see how different species are distributed

plt.figure(figsize=(6,4))

sns.scatterplot(
    data=df,
    x="sepal_length",
    y="sepal_width",
    hue="species"
)

plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
# Histograms show distribution of each numerical feature
# Helps us understand how values are spread

df.hist(figsize=(8,6), bins=15)

plt.suptitle("Feature Distributions")
plt.show()
```

```python
# Box plot helps detect outliers and spread of data
# Each feature is visualized to see median and quartiles

plt.figure(figsize=(6,4))

sns.boxplot(data=df)

plt.title("Box Plot of Iris Features")
plt.show()
```

```python
.Explanation of Results and Final Insights

From the exploratory data analysis of the Iris dataset:

- The dataset contains 150 samples and 4 numerical features.
- There are no missing values, so no data cleaning was required.
- The scatter plot shows that different species form visible clusters, especially Setosa.
- Histograms show that petal measurements have clearer separation than sepal measurements.
- Box plots indicate that only a few minor outliers are present in the dataset.

.Final Insight:
Petal length and petal width are the most useful features for distinguishing between Iris species.
```
